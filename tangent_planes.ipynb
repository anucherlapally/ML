Why must the tangent object be a plane and not something curved?

When we infinitely zoom in on a 3D surface, we come up on a flat surface which resembles a plane. So the point on the surface at which the surface is localized is a tangent plane.

Because the function is differentiable, all higher-order curvature terms vanish faster than linear ones, leaving only a first-order (linear) object — a plane.

Why do partial derivatives appear as coefficients?

When we take the equation of a plane and apply limit on the change tending to zero, we are left with only the coefficients of the linear terms, these are the partial derivatives.

Local expansion:

f(x0+Δx,y0+Δy) = f(x0,y0) + AΔx + BΔy + higher order terms

Those coefficients A,B must equal the partial derivatives.

Why is the dot product unavoidable here?

When we take the equation of a plane -
        z = Ax + By

Applying change - z-z0 = A(x - x0) + B(y - y0)
                - deltaZ ~ A(deltaX) + B(deltaY)

    here the change is the vector v with components (deltaX, deltaY)
    considering A and B as the partial derivatives of the gradient f
    rewriting the equation as deltaZ ~ nablaF . v

and so the dot products appears in the equation.

What information is missing from partial derivatives alone that the total derivative provides?

partial derivatives provide information of the change in f with respect to one axis keeping the others constant. total derivative provide the entire rate of change.

Partial derivatives are directional derivatives along coordinate axes.
The total derivative gives the linear map that predicts change in any direction.
