```python
import numpy as np
```

```python
n = 100
d = 2
np.random.seed(42)
X = np.random.randn(n, d)

w_true = np.array([2.0, -3.0])

noise = 0.1 * np.random.randn(n)

y = X @ w_true + noise
```

```python
def gradient(w0):
    error = X @ w0 - y
    grad = 2 * np.transpose(X) @ (error)
    return grad
```

```python
# gradient descent iteration
def grad_iteration(learning_rate, iterations=250):
    loss = []
    w0 = np.array([0, 0])
    for _ in range(1, iterations):
        w1 = w0 - learning_rate * (gradient(w0))
        w0 = w1
        loss.append(np.linalg.norm(X @ w0 - y)**2)

    return (loss, w1)
```

```python
eigs = np.linalg.eigvals(X.T @ X)
print(eigs)
```

```python
#Optimal learning_rate < 1/eigenvalue_max

learning_rate =  0.5 * 1 / np.max(eigs)
print(learning_rate)
```

```python
loss, final_weight = grad_iteration(learning_rate)
print(loss)
print(final_weight)
```

```python
w_star = np.linalg.inv(X.T @ X) @ X.T @ y
print(w_star)
```

```python
print(np.linalg.norm(final_weight - w_star))
```

```python
lambda_max = np.max(eigs)
lambda_min = np.min(eigs)
eta_opt = 2 / (lambda_max + lambda_min)
print(eta_opt)
```

```python
#Ill-conditioning simulation
np.random.seed(42)

n = 100

x1 = np.random.randn(n)
x2 = x1 + 0.001 * np.random.randn(n)   # almost linear combination

X = np.column_stack([x1, x2])

w_true = np.array([2.0, -3.0])
noise = 0.1 * np.random.randn(n)

y = X @ w_true + noise
```

```python
eigs = np.linalg.eigvals(X.T @ X)
print(eigs)
```

```python
print("Condition number:", np.max(eigs) / np.min(eigs))
```

```python
Identity = np.eye(2, 2)
H_ridge = 2 * (X.T @ X + Identity)
ridge_eigs = np.linalg.eigvals(H_ridge)
print(ridge_eigs)
print("Condition number: ", np.max(ridge_eigs) / np.min(ridge_eigs))
```

```python
eta_max = 1 / np.max(eigs)
eta_max_ridge = 1 / (np.max(ridge_eigs))
print("Max stable eta:", eta_max)
print("Ridge eta: ", eta_max_ridge)
```

```python
eta_opt = 0.5 * eta_max
loss, final_weight = grad_iteration(eta_opt, iterations=1000)
print(loss)
print(final_weight)
```

```python
def ridge(w0, regularization_factor=1):
    error = X @ w0 - y
    grad = 2 * np.transpose(X) @ (error) + (2 * regularization_factor * w0)
    return grad
```

```python
# ridge iteration
def ridge_iteration(learning_rate, iterations):
    loss = []
    w0 = np.array([0, 0])
    for _ in range(1, iterations):
        w1 = w0 - learning_rate * (ridge(w0))
        w0 = w1
        loss.append(np.linalg.norm(X @ w0 - y)**2)

    return (loss, w1)
```

```python
eta_opt = 0.5 * eta_max_ridge
loss, final_weight = ridge_iteration(eta_opt, 1000)
print(loss)
print(final_weight)
```
