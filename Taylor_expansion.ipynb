Why is curvature in multiple dimensions direction-dependent, but not in 1D?

In 1D, there is only one direction → curvature is a single number.
In multiple dimensions, there are infinitely many directions you can move.

Curvature can be:
positive in one direction
negative in another
zero in yet another

That’s why curvature becomes direction-dependent.

What information does the Hessian contain that the gradient does not?

Gradient provides information on the slopes in the direction of x and y. Hessian provides curvature as well, which is the effect of x on y, etc..

If the gradient at a point is zero, why is the Hessian still essential?

If the gradient at a point is zero, that would imply that the slope is flat, however as we move up curvature might still exist so Hessian is still essential.
