The slope of the hill if you walk purely east–west, without drifting north or south. It answers if I walk in the east-west direction, how much does does my height change?

Can a surface be flat in the x-direction but steep in the y-direction at the same point?

YES — absolutely

A surface can be flat in the x-direction, steep in the y-direction at the same point.

Partial derivatives measure independent directional behavior

Freezing y does not eliminate steepness in y — it isolates directions.

Why are partial derivatives still just ordinary derivatives in disguise?

Though partial derivatives take in multi-variable cases, we consider one effect at a time freezing all the other variables. Since all the other variables are now treated as constants, it's just ordinary derivative.
