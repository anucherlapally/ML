What does it mean for vectors to be linearly independent geometrically?
When vectors are said to be linearly indepenent geometrically, it means that they have no influence over other vectors in that column space. There is no redundancy because no other vector or combination of vectors can be used to describe any other vector.

Why does linear dependence destroy uniqueness in least squares?
In least squares we measure the minimum distance from 'b' to the closest projection on 'A', which is 'Ax', if there's linear dependence then there could be multiple 'x' that satisfy the equation and that destroys uniqueness.

Why is orthogonality central to least squares?
For least squares, we measure the distance from 'b' to the point closest in the column space, by definition the point closest has to be the point in the column space that's perpendicular to 'b'. Hence orthogonality is central to least squares

Eigenvector represent the directions that are squashed. These vectors live in the null space. 

Why are eigenvectors special directions?
Eigenvectors are special directions because they do not change in direction. After applying transformations, the vectors end up on the same line, scaled.
