Why must derivatives multiply in the chain rule rather than add?

Give a geometric explanation

Applying chain rule on how changes in 'x' affect 'y' -

g(x+h) ~ g(h) + g'(x)*h
deltaU ~ g'(x) * h

f(u + deltaU) ~ f(u) + f'(u) * deltaU
deltaY = f'(u) * deltaU

deltaY = f'(u) * g'(x) * h

deltaY / h = f'(u) * g'(x)

dy/dx = dy/du * du/dx

we see that terms result in multiplying rather than adding up.
