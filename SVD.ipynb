Why rank = number of non-zero singular values
Why null space corresponds to zero singular values
Why “low rank” means “low information”
Why truncating SVD is optimal, not heuristic

Because rank is all the directions that survive the transformation and so is the number of non-zero singular values

null space is all the directions that don't survive and so corresponds to zero singular values

low rank implies low non-zero singular values which says that the number of useful directions are less and hence low information. 
Low rank mathematically means few non-zero singular values
Low information practically means few large singular values

Truncating SVD is optimal because it reduces noise and leave the directions which are meaningful

Why condition number depends on singular values
Condition number is the ratio of max singular value to the min singular value. In the worst case scenario, the error for input x aligns with the smallest axis of the ellipse and the input is aligned to the larget axis of the ellipse.

Why small singular values are dangerous
Small singular values are dangerous because they have the potential to contribute to noise. The condition number will increase. 

Why inversion amplifies noise
Inversion amplifies noise because small singular number contribute more than the large singular numbers, they cause huge amplification and noise explodes

Why regularization improves stability
Regularization adds eigenvalue to the singular values, lifting them up, which decreases the condition number and stabilizes inversion.
