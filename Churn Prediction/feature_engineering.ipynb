```python
import pandas as pd
```

```python
df = pd.read_csv('churn_data.csv')
df = df.drop(columns='customerID')
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TechSupport'].unique()
```

```python
#Binary categorical columns
binary_cols = [col for col in df.columns if df[col].nunique() == 2]

for col in binary_cols:
    df[col] = df[col].astype("category").cat.codes
```

```python
from sklearn.preprocessing import OneHotEncoder

cols_to_encode = ['Contract','InternetService','PaymentMethod','MultipleLines', 'TechSupport', 'DeviceProtection', 'OnlineBackup', 'OnlineSecurity']

encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
encoded_array = encoder.fit_transform(df[cols_to_encode])

encoded_cols = encoder.get_feature_names_out(cols_to_encode)
df_encoded = pd.DataFrame(encoded_array, columns=encoded_cols)

encoded_df = pd.concat([df.drop(columns = cols_to_encode), df_encoded], axis=1)
encoded_df.head()
```
